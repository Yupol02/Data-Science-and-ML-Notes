## 🛠️ Bölüm 1: Pandas Veri Yapıları — Veri Serileri (Series) ve Vektörize Matematik

---

### 📖 Markdown Hücresi: Teorik Altyapı ve Endüstri Pratikleri

Pandas Veri Serileri (pd.Series), NumPy dizileri (arrays) üzerine inşa edilmiş, tek boyutlu ve etiketli (indexed) veri yapılardır. Python listelerinden en büyük farkı, her elemanın bir indeks etiketine sahip olması ve Hafıza Performansı (Vektörizasyon) sunmasıdır.

**Kritik Mühendislik Notu (Index Alignment):** İki seriyi topladığımızda veya işleme soktuğumuzda Pandas, elemanların sırasına değil, indeks etiketlerine bakar. Eğer indeksler eşleşmiyorsa, veri kaybını önlemek adına o satıra NaN (Not a Number) atar.

---

```python
import pandas as pd
import numpy as np 
```

```python
# 1.1. Sözlük (Dictionary) Üzerinden Seri Oluşturma
# Anahtarlar (Keys) otomatik olarak indeks etiketlerine dönüşür.
grades = {"Dennis": 50, "Linus": 60, "Ada": 30}
grades_series = pd.Series(grades)
print("Notlar Serisi:\n",grades_series)
```

```python
# 1.2. Explicit (Açık) İndeks Tanımlayarak Seri Oluşturma
# Veri ve indeks listelerinin uzunlukları birebir eşleşmelidir.
contest_result = pd.Series(data=[10, 5, 100], index=["Dennis", "Linus", "Ada"])
contest_result2 = pd.Series(data=[20, 50, 10], index=["Dennis", "Linus", "Ada"])
```

```python
# 1.3. Etiket Tabanlı Veri Erişimi (Label-Based Indexing)
# O(1) zaman karmaşıklığı (Time Complexity) ile çalışır, hash-map mantığındadır.
dennis_score1 = contest_result["Dennis"]
dennis_score2 = contest_result2["Dennis"]
print(f"Dennis'ın 1. Yarışma Skoru: {dennis_score1}, 2. Yarışma Skoru: {dennis_score2}\n")
```

```python
# 1.4. Vektörize Matematiksel İşlemler
# Pandas, döngü (for loop) kullanmadan element-wise (öğe bazlı) optimizasyon yapar.
final_result = contest_result + contest_result2
subtraction_result = contest_result - contest_result2
division_result = contest_result / contest_result2

print("Toplam Skorlar:\n", final_result)
```

```python
# 1.5. Kritik Durum: İndeks Uyuşmazlığı ve Otomatik Hizalama (Index Alignment)
# Farklı indekslere sahip seriler işleme sokulduğunda kesişmeyen indeksler NaN döner.
different_series = pd.Series([20, 30, 40, 50], index=["a", "b", "c", "d"])
different_series2 = pd.Series([10, 5, 3, 1], index=["a", "c", "f", "g"])

# Professional Tip: NaN oluşmasını engellemek için .add(), .sub() gibi fonksiyonlar 
# 'fill_value' parametresi ile kullanılmalıdır: different_series.add(different_series2, fill_value=0)
mismatched_sum = different_series + different_series2
print("\nUyuşmaz İndeks Toplam Sonucu (NaN Analizi):\n", mismatched_sum)
```

## 📊 Bölüm 2: DataFrame Anatomisi — İndeksleme, Slicing ve Çok Boyutlu Veri Yönetimi

--- 

### 📖 Markdown Hücresi: Veri Seçimi ve Bellek Yönetimi
DataFrame, satır ve sütunlardan oluşan iki boyutlu, etiketli bir veri yapısıdır. Veri manipülasyonunda en sık yapılan hata .loc ve .iloc kavramlarının karıştırılmasıdır:

* .loc[]: Etiket (Label) tabanlı seçim yapar.

* .iloc[]: Tam sayı (Integer position) tabanlı, yani 0 tabanlı indeks konumuna göre seçim yapar.

---

**Kritik Not (Slicing Farkı):** Standart Python listelerinde [0:5] dilimlemesi 5. indeksi **dahil etmezken**, Pandas .loc["A":"F"] veya .loc[0:5] işleminde üst sınırı **dahil eder**. Ayrıca, veri manipülasyonlarında kalıcı değişiklikler için **inplace=True** parametresi dikkatli kullanılmalıdır; modern Pandas mimarisinde, zincirleme atamaları (chained assignment) önlemek için df = df.drop(...) yöntemi inplace=True kullanımına göre daha çok teşvik edilmektedir.

---

```python
# 2.1. Standart Normal Dağılımdan Rastgele Matris Üretimi ve DataFrame Dönüşümü
np.random.seed(42)  # Sonuçların tekrarlanabilirliği (Reproducibility) için seed sabitlendi.
raw_data = np.random.randn(4, 3)
data_frame = pd.DataFrame(raw_data)

# Veri Tiplerinin İncelenmesi
# DataFrame'in tek bir sütunu her zaman bir pd.Series nesnesidir.
print(f"DataFrame Tipi: {type(data_frame)}")
print(f"Sütun Tipi: {type(data_frame[0])}\n")
```

```python
# 2.2. Özelleştirilmiş Şema (Schema) ile DataFrame Oluşturma
new_df = pd.DataFrame(
    data=raw_data,
    index=["Dennis", "Linus", "Ada", "Alan"],
    columns=["Salary", "Age", "Seniority"]
)
print("Orijinal Veri Çerçevesi:\n", new_df, "\n")
```

```python
# 2.3. İleri Düzey İndeksleme ve Veri Seçimi (Indexing & Slicing)
print("Etiket ile Satır Seçimi (.loc):\n", new_df.loc["Dennis"])
print("\nKonum ile Satır Seçimi (.iloc):\n", new_df.iloc[0])
print("\nTüm Satırlar, İlk Sütun Seçimi (.iloc[:, 0]):\n", new_df.iloc[:, 0])
print("\nTek Sütun Seçimi (Series döner):\n", new_df["Age"])
print("\nÇoklu Sütun Seçimi (DataFrame döner):\n", new_df[["Age", "Salary"]])
```

```python
# 2.4. Sütun ve Satır Mutasyonları (Feature Engineering)
new_df["Ekstra"] = 10  # Broadcaster özelliği ile tüm satırlara 10 atanır.

# Bellekte Kalıcı Silme Operasyonu (inplace=True)
# axis=1 sütunları, axis=0 satırları temsil eder.
new_df.drop("Ekstra", axis=1, inplace=True) 

new_df["Alan"] = 10
new_df.drop("Alan", axis=1, inplace=True)
new_df.drop("Alan", axis=0, inplace=True)  # Satır silme
```

```python
# 2.5. Hücre Bazlı Değer Atama
new_df.loc["Dennis", "Salary"] = 100
print("\nGüncellenmiş Veri Çerçevesi:\n", new_df)
```

```python
# 2.6. Boole Maskeleme (Filtering & Boolean Masking)
# Vektörize filtreleme işlemleri, büyük veri setlerinde 'query' optimizasyonu sağlar.
boolean_mask = new_df > 0
filtered_df = new_df[new_df > 0]  # Koşulu sağlamayan hücreler NaN olur.
print(filtered_df)
```

```python
# 2.7. İndeks Manipülasyonları
new_df.reset_index(inplace=True)  # Eski indeks 'index' adıyla sütuna dönüşür.
new_df["NewIndex"] = ["Denn", "Lin", "Ala"]
new_df.set_index("NewIndex", inplace=True)  # Yeni sütun indeks haline getirilir.
```

```python
# 2.8. Hiyerarşik İndeksleme (Multi-Index / Advanced Structuring)
# Karmaşık verileri (örn: Zaman serileri, panel veriler) modellemek için kullanılır.
first_index = ["Simpson", "Simpson", "Simpson", "South Park", "South Park", "South Park"]
inner_index = ["Homer", "Bart", "Marge", "Cartman", "Kenney", "Kyle"]
```

```python
zipped_index = list(zip(first_index, inner_index))
multi_index = pd.MultiIndex.from_tuples(zipped_index)

sample_values = np.ones((6, 2))
big_df = pd.DataFrame(sample_values, index=multi_index, columns=["Age", "Salary"])
print("\nÇoklu İndeksli DataFrame:\n", big_df)
```

```python
# Multi-Index Veri Erişimi (Cross-Section Extraction)
bart_data = big_df.loc["Simpson"].loc["Bart"]
print("\nSimpson -> Bart Bilgileri:\n", bart_data)
```

## 🔀 Bölüm 3: Veri Birleştirme Stratejileri — Concat ve İlişkisel Merge (Join) İşlemleri
---

### 📖 Markdown Hücresi: Birleştirme Algoritmaları
Veri bilimi projelerinde veriler genellikle farklı kaynaklardan (farklı CSV'ler, SQL tabloları) parçalı olarak gelir. Bunları birleştirmek için iki ana strateji vardır:

1. **pd.concat():** Verileri dikey (satır bazlı) ya da yatay (sütun bazlı) olarak yapıştırır/ekler. ignore_index=True, yeni oluşan yapıda indekslerin sıfırdan sıralanmasını sağlar.

2. **pd.merge():** SQL'deki JOIN mantığı gibi çalışır. Belirlenen bir ortak anahtar sütun üzerinden (on="Key") ilişkisel birleştirme yapar:
    * **Inner:** Sadece her iki tabloda da ortak olan anahtarları alır.
    * **Outer:** Tüm anahtarları alır, eşleşmeyen yerleri NaN ile doldurur.
    * **Left / Right:** Sol veya sağ tablodaki tüm verileri korur, karşı taraftan eşleşenleri getirir.

---

```python
# Örnek veri setlerinin oluşturulması (Simülasyon)
df1 = pd.DataFrame({"A": ["A0", "A1"], "B": ["B0", "B1"]})
df2 = pd.DataFrame({"A": ["A2", "A3"], "B": ["B2", "B3"]})
df_merge1 = pd.DataFrame({"Employee_ID": [1, 2, 3], "Name": ["Alice", "Bob", "Charlie"]})
df_merge2 = pd.DataFrame({"Employee_ID": [2, 3, 4], "Salary": [50000, 60000, 70000]})
```

```python
# 3.1. Dikey Birleştirme (Concatenation)
# ignore_index=True veri indekslerinin çakışmasını engeller ve sürekli bir indeks oluşturur.
df_concat = pd.concat([df1, df2], ignore_index=True)
print("Dikey Birleştirme Sonucu (Concat):\n", df_concat, "\n")
```

```python
# 3.2. İlişkisel Birleştirmeler (SQL-like Merges)
# Outer Join: Tüm çalışanları ve maaşları getirir (Union)
df_merged_outer = pd.merge(df_merge1, df_merge2, on="Employee_ID", how="outer")
```

```python
# Left Join: İlk tablodaki (df_merge1) tüm çalışanları korur
df_merge_left = pd.merge(df_merge1, df_merge2, on="Employee_ID", how="left")
```

```python
# Right Join: İkinci tablodaki (df_merge2) tüm maaş kayıtlarını korur
df_merge_right = pd.merge(df_merge1, df_merge2, on="Employee_ID", how="right")
```

```python
# Inner Join: Sadece her iki tabloda da ID'si bulunan kayıtları getirir (Intersection)
df_merged_inner = pd.merge(df_merge1, df_merge2, on="Employee_ID", how="inner")

print("Inner Join (Kesişim) Sonucu:\n", df_merged_inner)
```

## ⚡ Bölüm 4: Gelişmiş Fonksiyon Uygulamaları — Apply ve Lambda ile Veri Manipülasyonu

---
### 📖 Markdown Hücresi: Fonksiyonel Programlama ve Performans İpuçları
Pandas'ta **.apply()** metodu, özel bir fonksiyonu veri çerçevesinin satır veya sütunları üzerinde çalıştırmaya olanak tanır.
* df["Sütun"].apply(func): Tek bir sütundaki elemanları teker teker fonksiyona sokar (Element-wise).
* df.apply(func, axis=1): Fonksiyona tüm satırı (row) bir bütün olarak gönderir. Satırlar arası mantıksal karşılaştırmalar için idealdir.

**Kritik Not (Performans Uyarısı):** .apply() fonksiyonu, arkada gizli bir for döngüsü çalıştırır ve saf Python hızında işlem yapar. Milyonlarca satırlık büyük veri setlerinde apply yerine Pandas'ın built-in vektörize fonksiyonları veya NumPy vektörizasyonları tercih edilmelidir. Küçük ve orta ölçekli verilerde ise kod okunabilirliği için mükemmel bir çözümdür.

---

```python
# Örnek Veri Seti Simülasyonu (8-apply_function_data.csv içeriği)
data_apply = {
    "Name": ["John_Doe", "Anna_Smith", "Peter_Jones"],
    "Salary": [45000, 65000, 120000],
    "Experience": [5, 12, 8],
    "Performance_Score": [3, 4, 2]
}
df = pd.DataFrame(data_apply)
```

```python
# 4.1. Veri Seti Keşif Adımları (Exploratory Data Analysis - EDA)
print("--- Veri Seti Özeti ---")
print(df.describe())  # Sayısal sütunların istatistiksel dağılımı
print("\n--- Yapısal Bilgiler ---")
df.info()             # Veri tipleri ve eksik değer kontrolü
```

```python
# 4.2. Tek Sütun Üzerinde Apply Kullanımı
def salary_category(salary):
    """Maaş değerine göre kategorik etiket döner."""
    if salary < 50000:
        return "Low"
    elif 50000 <= salary < 80000:
        return "Medium"
    else:
        return "High"

df["Salary_Category"] = df["Salary"].apply(salary_category)
```

```python
# 4.3. Satır Bazlı (Multi-column) Apply Kullanımı (axis=1)
def adjust_performance(row):
    """Deneyimi 10 yıldan fazla olan personelin performans skorunu optimize eder."""
    if row["Experience"] > 10:
        return row["Performance_Score"] + 1
    return row["Performance_Score"]

df["Adjusted_Performance"] = df.apply(adjust_performance, axis=1)
```

```python
# 4.4. Anonim Fonksiyonlar (Lambda Expressions) ile String Manipülasyonu
# Tek satırlık, kalıcı olmayan dönüşümler için ideal endüstri pratiğidir.
df["Formatted_Name"] = df["Name"].apply(lambda x: x.replace("_", " "))

print("\nFonksiyon Uygulamaları Sonrası DataFrame:\n", df.head())
```

## 📈 Bölüm 5: Eksik Veri Yönetimi ve İleri Düzey Agregasyon (Groupby) Analitiği
---
## 📖 Markdown Hücresi: Split-Apply-Combine Teorisi
Veri analizinin en güçlü araçlarından biri **groupby** mekanizmasıdır. Bu yapı veri bilimi literatüründe Split-Apply-Combine (Parçala-Uygula-Birleştir) mimarisi olarak bilinir:

1. Split: Veri, belirlenen kategorik sütuna (örn: Departman, Şehir) göre gruplara ayrılır.

2. Apply: Her gruba kendi içinde istatistiksel bir fonksiyon (ortalama, toplam, sayma) uygulanır.

3. Combine: Sonuçlar tek bir özet tablo olarak birleştirilir.

**Eksik Veri (Missing Value) Yönetimi:** Gerçek hayat veri setleri boşluklarla (NaN, None) doludur. .isna().sum() komutu, veri setindeki eksik verilerin sütun bazlı haritasını çıkararak veri temizleme (data cleaning) stratejimizi belirlememizi sağlar.

---

```python
# Örnek Hava Durumu Verisi Simülasyonu (6-weather.xlsx içeriği)
weather_data = {
    "Istanbul": [22, 24, np.nan, 25, 21, 23],
    "New York": [15, np.nan, 18, 19, 14, 16],
    "London": [12, 13, 14, 11, np.nan, 15]
}
weather_df = pd.DataFrame(weather_data)
```
