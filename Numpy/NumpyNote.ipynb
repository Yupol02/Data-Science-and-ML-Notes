### Vektörel İşlemler (Element-wise Operations)
Python listelerinde + operatörü listeleri yan yana birleştirirken (concatenation), NumPy array'lerinde matematiksel işlemler karşılıklı elemanlar (element-wise) üzerinden yapılır.

---

```python
# Python listelerinde davranış
my_list1 = [1, 2]
my_list2 = [2, 3]
```

```python
# NumPy Array'lerinde davranış
my_npmy_array1 = np.array(my_list1)
my_npmy_array2 = np.array(my_list2)
```

```python
# Matematiksel operatörler doğrudan elemanlara uygulanır
print(my_npmy_array1 + my_npmy_array2)
print(my_npmy_array1 * my_npmy_array2) 
print(my_npmy_array1 - my_npmy_array2) 
print(my_npmy_array1 / my_npmy_array2) 
```

### İndeksleme ve Parçalama (Slicing)
Tek boyutlu NumPy dizilerinde indeksleme, klasik Python listeleriyle birebir aynı mantıkta çalışır.

---

```python
np_array = np.arange(0, 10)

np_array[-1]     # Son elemanı getirir (9)
np_array[0]      # İlk elemanı getirir (0)
```

```python
# Slicing: [başlangıç : bitiş : adım]
np_array[1:10:2] # 1. indeksten başla, 10'a kadar 2'şer atlayarak git
np_array[::-1]   # Array'i tersine çevirir (Reverse işlemi)
```

### Matrisler ve Skaler İşlemler
Makine öğreniminde veriler genellikle 2 boyutlu (satır ve sütunlar - tabular veri) veya daha çok boyutlu tensörler olarak temsil edilir.

---

```python
# İç içe listelerden matris oluşturma
my_matrix = [[5, 10], [15, 20]]
my_matrix[1][0] # 2. listenin ilk elemanına (15) erişim
```

```python
# Matrisi NumPy dizisine çevirme
numpy_matrix = np.array([[5, 10], [15, 20]])

# Toplam ve Özel Matrisler
numpy_matrix.sum()        # Matristeki tüm değerleri toplar (50)
np.ones((3, 2))           # Sadece 1'lerden oluşan 3 satır, 2 sütunlu matris
np.zeros((3, 5))          # Sadece 0'lardan oluşan 3 satır, 5 sütunlu matris
np.random.random((4, 4))  # 0.0 ile 1.0 arasında rastgele değerlerle 4x4 matris
```

```python
# Skaler İşlemler (Tekil bir sayıyı matrise uygulama)
first_array = np.array([[10, 20], [30, 40]])
first_array * 3 # Tüm elemanları 3 ile çarpar
first_array / 2 # Tüm elemanları 2'ye böler
first_array + 5 # Tüm elemanlara 5 ekler
```

### Broadcasting ve Matris Çarpımı (Dot Product)
Broadcasting: İki farklı boyuttaki array işleme girdiğinde, NumPy küçük olan boyutu mantıksal olarak "genişleterek" işleme uygun hale getirir.

---

```python
first_array = np.array([[10, 20], [30, 40]]) # shape: (2, 2)
third_array = np.array([[10], [20]])         # shape: (2, 1)

# third_array, (2,2) boyutuna sanal olarak yayılarak (broadcast) toplanır.
third_array + first_array
```

---
### Dot Product (Matris Çarpımı):
Makine öğrenimindeki yapay sinir ağlarının temel taşıdır. Karşılıklı elemanların çarpılıp toplanmasına dayanır.Kural: 1. matrisin sütun (column) sayısı ile 2. matrisin satır (row) sayısı **EŞİT** olmalıdır. Sonuç matrisinin boyutu: (1. matrisin satır sayısı) x (2. matrisin sütun sayısı) olur.Matematiksel gösterimi: $(m \times n) \cdot (n \times p) = (m \times p)$

```python
first_matrix = np.array([[10, 20, 30]]) # 1 satır, 3 sütun -> shape (1, 3)
second_matrix = np.array([[2, 3], 
                          [2, 3], 
                          [2, 3]])      # 3 satır, 2 sütun -> shape (3, 2)

# (1, 3) . (3, 2) = (1, 2) boyutunda sonuç beklenir.
result_matrix = first_matrix.dot(second_matrix)
print(result_matrix.shape)
```

### Koşullu Seçim (Boolean Masking)
Veri setindeki istenmeyen veya belirli şartlara uyan verileri filtrelemek (örneğin anomalileri çıkarmak) için sıklıkla kullanılır.

---

```python
new_array = np.random.randint(1, 100, 20) # 1-100 arası 20 sayı

# Her bir eleman için koşulu test eder ve True/False array'i döndürür
print(new_array > 25) 

# Köşeli parantez içine koşulu yazdığımızda, sadece True olan değerleri filtreler
filtrelenmis_veri = new_array[new_array > 25]
```

### Şekillendirme (Reshape) ve Transpoz (Transpose)
Array boyutlarını veriye zarar vermeden (toplam eleman sayısını sabit tutarak) yeniden düzenlemek, modellerin girdi (input) şekillerini ayarlarken kritik bir adımdır.

--- 

```python
matrix_array = np.array([[10, 20], [20, 30], [30, 40]]) # shape: (3, 2)

# Transpoze: Satırları sütun, sütunları satır yapar.
matrix_array.transpose() 
matrix_array.T # Yukarıdaki işlemle tamamen aynıdır (Kısa yazım)

# Reshape: Array'in şeklini yeniden boyutlandırma
random_array = np.random.random((6, 1)) # 6 elemanlı (6, 1) boyutlu vektör

# Toplam eleman sayısı (6x1=6) sabit kaldığı sürece farklı şekillere sokulabilir:
random_array.reshape(2, 3) # 2 satır, 3 sütunlu matrise çevirir (2x3=6)
random_array.reshape(3, 2) # 3 satır, 2 sütunlu matrise çevirir (3x2=6)
```
